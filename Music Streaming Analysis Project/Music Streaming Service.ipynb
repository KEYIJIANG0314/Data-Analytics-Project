```python
df_1 = df_n.loc[:,'Q4':'Q5_6']
df_1 = df_1.drop(['Q4_10_TEXT'],axis=1)
df_1 = df_1.dropna()
```

```python
df_1 = df_1.rename(columns={"Q4": "Current Music Streaming Service", "Q5_1": "Recommend by A Friend",
                     "Q5_2": "Price", "Q5_3": "Family plan",
                    "Q5_4": "Advertisements", "Q5_5": "Music Recommendations",
                    "Q5_6": "Collaborative Playlist Features"})
```

```python
labels = df_1['Current Music Streaming Service'].unique().tolist()
df_1['Current Music Streaming Service'] = df_1['Current Music Streaming Service'].apply(lambda x: labels.index(x))
```

```python
plt.figure(figsize=(10, 10))
sns.heatmap(df_1.corr(), vmax=.8, linewidths=0.01,square=True,annot=True,cmap='YlGnBu' ,linecolor="white")
plt.title('Correlation between Influence to selection')
```

```python
from sklearn.linear_model import LinearRegression
from sklearn.metrics import r2_score
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import (StandardScaler, PolynomialFeatures)
mlr = LinearRegression()
X = df_1[['Recommend by A Friend', 'Price', 'Family plan', 'Advertisements', 'Music Recommendations', 'Collaborative Playlist Features']]
y = df_1[["Current Music Streaming Service"]]
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.1, random_state=42) #split into train and test parts
s = StandardScaler()
X_train_sm = s.fit_transform(X_train)
mlr.fit(X_train_sm, y_train)
X_test_sm = s.transform(X_test)
y_pred_sm = mlr.predict(X_test_sm)
```

```python
print(f'R2 score is {r2_score(y_pred_sm,y_test)}')
print("Residual sum of squares (MSE): %.2f" % np.mean((y_pred_sm - y_test) ** 2))
print(X.columns)
print(f'Co-efficients are {mlr.coef_}')
print(f'Intercept is {mlr.intercept_}')
```

```python
df_2 = df_n.loc[:,['Q8_3','Q9_3']]
df_2 = df_2.dropna()
df_2.shape
```

```python
from sklearn.linear_model import LinearRegression
from sklearn.metrics import r2_score
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import (StandardScaler, PolynomialFeatures)
slr = LinearRegression()
X = df_2[['Q9_3']]
y = df_2[["Q8_3"]]
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.1, random_state=42)
```

```python
s = StandardScaler()
X_train_s = s.fit_transform(X_train)
slr.fit(X_train_s, y_train)
X_test_s = s.transform(X_test)
y_pred = slr.predict(X_test_s)
print(f'R2 score is {r2_score(y_pred,y_test)}')
print(f'Co-efficients are {slr.coef_}')
print(f'Intercept is {slr.intercept_}')
```

```python
plt.figure(figsize = (15,9))
sns.regplot(X_test_s,y_pred,scatter_kws={"color": "red"},line_kws={"color": "green"})
plt.xlabel('Satisfaction')
plt.ylabel('Importance')
```
